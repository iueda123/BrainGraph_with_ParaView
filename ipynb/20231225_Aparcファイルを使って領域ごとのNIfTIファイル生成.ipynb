### Create ROIs

```bash
#
# Create Rois
#


RoiOutputFolder=${DefaultWd}/nii.gz/Rois_${AtlasName}
CenteroidOuputFolder=${FSAVERAGE_TEMPLATE_PATH}/centeroids
mkdir -p ${CenteroidOuputFolder}

echo "\${AtlasName}=${AtlasName}"

if [[ -d ${RoiOutputFolder} ]]; then
#if [[ ! -d ${RoiOutputFolder} ]]; then

    echo "The output folder '${RoiOutputFolder}' already exists. Please check for its existance."
    echo "If you want to recreate the folder, please run the following command: "
    echo ""
    echo "    rm -r \${${RoiOutputFolder}}"
    echo ""

else

    mkdir -p ${RoiOutputFolder}

    # Specify TotalRegionCount
    if [ ${AtlasName} = "HCPMMP1" ]; then
        TotalRegionCount=$((360+16+2+1))
    elif [ ${AtlasName} = "NSPN500" ]; then  # modified修正未
        TotalRegionCount=$((308+5+16+2+1))
    elif [ ${AtlasName} = "DK" ]; then
        TotalRegionCount=$((68+5+16+2+1))
    elif [ ${AtlasName} = "EK" ]; then  # modified修正未
        TotalRegionCount=$((86+5+16+2+1))
    elif [ ${AtlasName} = "Yeo17" ]; then
        TotalRegionCount=$((34+5+16+2+1))
    elif [ ${AtlasName} = "Yeo7" ]; then
        TotalRegionCount=$((14+5+16+2+1))
    else
        echo "ERROR: Unrecognized atlas name was specified."
        exit 1
    fi
    echo "\${TotalRegionCount}: ${TotalRegionCount}"

    line_num=0
    RoiNum=0
    RundomNumber=${RANDOM}
    cat ${RoiVal_and_RoiName_ListFile} | while read line
    do
        line_num=$((line_num + 1))
        echo "====== Now processing line number ${line_num} ======"


        if [[ -z "${line// }" ]]; then
            echo "The line seems to be empty line."
            # 注意！：改行文字は「Unix/Linux」でないと上手く空白行として捕捉できないので注意！
        elif [[ "${line// }" =~ ^#.+ ]]; then
            echo "The line seems to be comment line."
        else
            
            array=(${line// / })
            echo "        RoiVal=${array[0]}, RoiName=${array[1]}"
            RoiNum=$((RoiNum + 1))
            RoiNumStr=`printf "%03d" "${RoiNum}"`
            RoiVal=${array[0]}
            RoiName=${array[1]}

            echo "        Try to create ${RoiOutputFolder}/${RoiNumStr}_${RoiName}.nii.gz"
            
            SubjectId=NoName
            
            ## Create Mask (fslコマンドのみを使った場合)
            #fslmaths ${SnspcAparcFile} -thr ${RoiVal} /tmp/${SubjectId}_tmp1.nii.gz
            #fslmaths /tmp/${SubjectId}_tmp1.nii.gz -uthr ${RoiVal} /tmp/${RoiNumStr}_${RoiName}.nii.gz

            # Create Mask (MRtrix3コマンドも使った場合)
            mrconvert -force  ${AparcFile} /tmp/${SubjectId}_${RundomNumber}_tmp1.mif # mif化
            mrcalc -force /tmp/${SubjectId}_${RundomNumber}_tmp1.mif ${RoiVal} -eq /tmp/${SubjectId}_${RundomNumber}_tmp2.mif # 2値（0, 1）化
            
            mrconvert -force /tmp/${SubjectId}_${RundomNumber}_tmp2.mif ${RoiOutputFolder}/${RoiNumStr}_${RoiName}.nii.gz

            # Calc Centeroid
            centeroid=`fslstats ${RoiOutputFolder}/${RoiNumStr}_${RoiName}.nii.gz -C`  #重心（Centeroid）の算出
            echo "        Centeroid of ${RoiName}: ${centeroid}"
            echo "${RoiNumStr} ${RoiName} ${centeroid}" >> ${CenteroidOuputFolder}/${AtlasName}.txt

        fi

        if [ ${RoiNum} -gt ${TotalRegionCount} ]; then
            break
        fi

        # For debug
        #if [ ${line_num} -eq 4 ]; then
        #    break
        #fi
        echo ""

    done

fi
```
