## Generate sample value table for 16 objects

```r
# Subcortical Regions and Lateral Ventricles
ScRLV <- c("L_thal", "R_thal", "L_caud", "R_caud", "L_put", "R_put", "L_pal", "R_pal", "L_hippo", "R_hippo", "L_amyg", "R_amyg", "L_accumb", "R_accumb", "L_LatVent", "R_LatVent")
length(ScRLV)
```

```r
SequentialValues_16 <- runif(n=16, min=0, max=1)
#SequentialValues_16
df_for_output <- data.frame(OBJECT=ScRLV, VALUE=SequentialValues_16)
df_for_output
output_fld_path <- "./ValueTables"
filename_prefix <- "ScRLV"
filename_body <- "SequentialValues_16"
output_file_path <- paste(output_fld_path, "/", filename_prefix, "_", filename_body, ".tsv", sep="")
write.table(df_for_output, output_file_path, quote=F, append=F, row.names=F, col.names=F, sep="\t")
cat(output_file_path, " was generated.", sep="")
```

```r
DivergingValues_16 <- runif(n=16, min=-1, max=1)
#DivergingValues_16
df_for_output <- data.frame(OBJECT=ScRLV, VALUE=DivergingValues_16)
df_for_output
output_fld_path <- "./ValueTables"
filename_prefix <- "ScRLV"
filename_body <- "DivergingValues_16"
output_file_path <- paste(output_fld_path, "/", filename_prefix, "_", filename_body, ".tsv", sep="")
write.table(df_for_output, output_file_path, quote=F, append=F, row.names=F, col.names=F, sep="\t")
cat(output_file_path, " was generated.", sep="")
```

```r
QualitativeValues_16 <- runif(n=16, min=1, max=12)
#QualitativeValues_16
df_for_output <- data.frame(OBJECT=ScRLV, VALUE=QualitativeValues_16)
df_for_output
output_fld_path <- "./ValueTables"
filename_prefix <- "ScRLV"
filename_body <- "QualitativeValues_16"
output_file_path <- paste(output_fld_path, "/", filename_prefix, "_", filename_body, ".tsv", sep="")
write.table(df_for_output, output_file_path, quote=F, append=F, row.names=F, col.names=F, sep="\t")
cat(output_file_path, " was generated.", sep="")
```
